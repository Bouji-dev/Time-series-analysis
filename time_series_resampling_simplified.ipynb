```python
import numpy as np
import pandas as pd
```

```python
date_rng = pd.date_range(start='2021-01-01', end='2021-02-10', freq='h')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0, 100, size=len(date_rng))
```

```python
df.head()
```

```python
df.set_index('date', inplace=True)
```

```python
df
```

```python
daily_df = df.resample('d').mean()
```

```python
daily_df.head()
```

```python
df_30m = df.resample('30min').ffill()
df_30m
```

```python
df_custom = df.resample('51h').mean()
df_custom
```

```python
df_weekly = df.resample('W').sum()
df_weekly
```

```python
df_monthly = df.resample('ME').sum()
df_monthly
```

```python
df_15min = df.resample('15min').bfill()
```

```python
df_15min
```

```python
df_combo = df.resample('30min').ffill().resample('d').mean()
```
