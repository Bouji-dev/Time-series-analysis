```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date_rang = pd.date_range(start='2022-01-01', end='2023-01-01', freq='D')
```

```python
date_rang
```

```python
data = {
    'date': date_rang,
    'values': np.linspace(10, 50, len(date_rang)) + np.random.normal(scale= 5, size= len(date_rang)) ,
    'column1': np.random.randint(1, 100, size=len(date_rang)),
    'column2': np.random.choice(['A', 'B', 'C'], size=len(date_rang))
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
plt.scatter(df['date'], df['values'], label='values', linestyle= '-', marker= 'o', color='b')
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
df.head()
```

```python
df['year'] = df['date'].dt.year
```
